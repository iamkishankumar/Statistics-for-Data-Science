### One Sample T-test

```python
from scipy.stats import ttest_1samp
import numpy as np 
import numpy.random
```

```python
data = np.random.normal(loc = 30,scale = 8,size = 100)
l = []
for i in data:
    l.append(int(round(i,0)))
```

```python
np.mean(l)
```

```python
np.median(l)
```

```python
len(l)
```

```python
## Lets take sample
l = np.array(l)
sample_size=10
age_sample=np.random.choice(l,sample_size)
```

```python
age_sample
```

```python
age_sample.mean()
```

```python
ttest,p_value= ttest_1samp(age_sample,l.mean())
```

```python
print(ttest,p_value)
```

```python
if p_value < 0.05:    # alpha value is 0.05 or 5%
    print(" we are rejecting null hypothesis")
else:
    print("we are accepting null hypothesis")
```

### Chi square test

```python
import seaborn as sns 
```

```python
df = sns.load_dataset("tips")
```

```python
df
```

**Observation Table** 

```python
dataset_table = pd.crosstab(df['sex'],df['smoker'])
print(dataset_table)
```

```python
dataset_table.values
```

```python
#Observed Values
Observed_Values = dataset_table.values 
print("Observed Values :-\n",Observed_Values)
```

Returns
-------
chi2 : float
    The test statistic.
p : float
    The p-value of the test
dof : int
    Degrees of freedom
expected : ndarray, same shape as `observed`
    The expected frequencies, based on the marginal sums of the table.


```python
import scipy.stats 
```

```python
val=scipy.stats.chi2_contingency(dataset_table)
```

```python
val
```

```python
if val[1]<0.05:
    print("reject Null , there is relationship between smoker and sex")
else:
    print("accept null , there is no relationship between smoker and sex")
```

```python
Expected_Values=val[3]
Expected_Values
```

```python
no_of_rows=len(dataset_table.iloc[0:2,0])
no_of_columns=len(dataset_table.iloc[0,0:2])
ddof=(no_of_rows-1)*(no_of_columns-1)
print("Degree of Freedom:-",ddof)
alpha = 0.05
```

```python
from scipy.stats import chi2
chi_square=sum([(o-e)**2./e for o,e in zip(Observed_Values,Expected_Values)])
chi_square_statistic=chi_square[0]+chi_square[1]
```

```python
chi_square_statistic
```

```python
critical_value=chi2.ppf(q=1-alpha,df=ddof)
print('critical_value:',critical_value)
```

```python
p_value=1-chi2.cdf(x=chi_square_statistic,df=ddof)
print('p-value:',p_value)
print('Significance level: ',alpha)
print('Degree of Freedom: ',ddof)
```

```python
if chi_square_statistic>=critical_value:
    print("Reject H0,There is a relationship between 2 categorical variables")
else:
    print("Retain H0,There is no relationship between 2 categorical variables")
    
if p_value<=alpha:
    print("Reject H0,There is a relationship between 2 categorical variables")
else:
    print("Retain H0,There is no relationship between 2 categorical variables")
```
